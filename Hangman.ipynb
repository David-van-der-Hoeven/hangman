```python
from random_word import RandomWords

def get_word():
    r = RandomWords()

    random_word = r.get_random_word()
    return random_word.lower()
```

```python
def replace_letter(guess, word_guess, word):
    word_list, guess_list = [x for x in word], [x for x in word_guess]
    for index, letter in enumerate(word_list):
        if letter == guess:
            guess_list[index] = guess
    return ''.join(guess_list) 
```

```python
def again():
    once_more = input("Would you like to play again? y/n ")
    if once_more == 'y':
        tasks()
    else:
        print('Goodbye')
```

```python
def get_guess(guessed_letters):
    guess = input('Guess a letter please: ').lower()
    if not (guess.isalpha()) & (len(guess) == 1):
        print("Only single letters please")
        guess = get_guess(guessed_letters)
    if guess in guessed_letters:
        print("You\'ve already guessed that letter")
        guess = get_guess(guessed_letters)
    return guess
```

```python
def tasks():
    guesses = 0
    word = get_word()
    word_guess = len(word) * '*'
    guessed_letters = []
    while guesses < 6:
        if not '*' in word_guess:
            print('You win!')
            print('You guessed ' + str(guesses) + ' times')
            again()
        
        print(hangmen[guesses])
        print('Guessed letters: ')
        print(guessed_letters)
        print(word_guess)
        guess = get_guess(guessed_letters)
        
        guessed_letters.append(guess)
        if guess in word:
            word_guess =  replace_letter(guess, word_guess, word)
            print('Good guess!')
        else:
            guesses += 1
            print('Nope')
    print(hangmen[6])
    print('GAME OVER')
    print(f'The word was {word}')
    again()
```
